```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("all_features_from_BonnUniversityData.csv")
```

```python
data
```

```python
del data['Unnamed: 0']
```

```python
data
```

Creating target vector:

```python
target=[]
for i in range(0,2300):
    target.append(0)
for i in range(0,2300):
    target.append(0)
for i in range(0,2300):
    target.append(1)
for i in range(0,2300):
    target.append(1)
for i in range(0,2300):
    target.append(1)
```

```python
target=pd.DataFrame(target,columns=['y'])
```

```python
target
```

## Feature selection 

```python
from sklearn.feature_selection import SelectKBest , mutual_info_classif
```

```python
selector = SelectKBest(mutual_info_classif,k=30)
```

```python
selector.fit(data,target.values.ravel())
```

Selected features:

```python
data.columns[selector.get_support()]
```

#### Creation of a dataframe that contains the selected features:

```python
selected_features=pd.DataFrame()
for i in data.columns[selector.get_support()]:
    selected_features[i]=data[i]
    
```

```python
selected_features
```

#### Separating data into training and testing sets:

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(selected_features,target,test_size=0.2,random_state=0)
```

#### Data standardization:

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

## K-nearest neighbors classifier

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix , roc_auc_score
from sklearn.metrics import classification_report
```

```python
leaf_size = list(range(1,20))# Par défaut c'est 30
n_neighbors = list(range(1,25))
p=[1,2] #p = 1 manhattan_distance/p=2 euclidean_distance
#Convert to dictionary
hyperparameters = dict(leaf_size=leaf_size, n_neighbors=n_neighbors, p=p)
#Create new KNN object
knn_2 = KNeighborsClassifier()
#Use GridSearch
clf = GridSearchCV(knn_2, hyperparameters, cv=5, n_jobs=-1)
#n_jobs: number of CPU'sfor execution.
#cv: number of folds of the cross validation

#Fit the model
best_model = clf.fit(X_train, y_train)
#Print The value of best Hyperparameters
print('Best leaf_size:', best_model.best_estimator_.get_params()['leaf_size'])
print('Best p:', best_model.best_estimator_.get_params()['p'])
print('Best n_neighbors:', best_model.best_estimator_.get_params()['n_neighbors'])
```

```python
knn = KNeighborsClassifier(n_neighbors=9,leaf_size=1,p=2)
knn_model = knn.fit(X_train, y_train.values.ravel())
y_pred_knn = knn_model.predict(X_test)
```

```python
print('Accuracy of K-NN classifier on training set: {:.2f}'
     .format(knn.score(X_train, y_train)))
print('Accuracy of K-NN classifier on test set: {:.2f}'
     .format(knn.score(X_test, y_test)))
```

```python

print(confusion_matrix(y_test, y_pred_knn))
```

```python
print(classification_report(y_test, y_pred_knn,target_names=['Healthy','Epileptic']))
```

```python
print("AUC score:",roc_auc_score(y_test,y_pred_knn))
```

## Artificial neural network

```python
import tensorflow as tf
import keras
import keras.backend as K
from keras.models import Sequential
from keras.layers import Dense,Activation,Dropout
from keras.utils.vis_utils import plot_model
import pydot
from keras.optimizers import SGD
```

### Data pre-processing

Loading raw data:

```python
data=pd.read_csv("bonn_university_data.csv")
```

```python
data.head()
```

```python
del data['time']
```

Splitting the signals into several windows:

```python
a=[]
for i in data.columns:
    a.append(np.array_split(data[i],23))
b=[]
for j in range(0,500):
    for k in range(0,23):
        b.append(a[j][k])
```

```python
for i in range(0,len(b)):
    b[i]=b[i].to_numpy()
```

Creating the vector target:

```python
# 0 = Healthy
# 1 = Epileptic
target=[]
for i in range(0,2300):
    target.append(0)
for i in range(0,2300):
    target.append(0)
for i in range(0,2300):
    target.append(1)
for i in range(0,2300):
    target.append(1)
for i in range(0,2300):
    target.append(1)
```

```python
target=np.array(target)
```

Adjustin the window sizer for each sample

```python
for i in range(0,len(b)):
    if b[i].shape[0]!=178:
        b[i]=np.delete(b[i],178)
```

Separating data into training and testing sets

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(b,target,test_size=0.2,random_state=0)
```

Data normalization:

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

Initialising the ANN

```python
model = Sequential()
```

Adding the input layer and the first hidden layer

```python
model.add(Dense(120,input_dim=178,activation='relu'))
model.add(Dropout(0.2))
```

Adding the second hidden layer

```python
model.add(Dense(80,activation='relu'))
model.add(Dropout(0.2))
```

Adding the output layer

```python
model.add(Dense(1,activation='sigmoid'))
```

Model summary:

```python
model.summary()
```

Compiling the ANN:

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

Fitting the ANN to the Training set

```python
model.fit(X_train, y_train,epochs=500)
```

Model evaluation:

```python
print("Accuracy of the ANN on train set:",model.evaluate(X_train,y_train))
print("Accuracy of the ANN on test set:",model.evaluate(X_test,y_test))
```

```python
ytest = model.predict_classes(X_test)
from sklearn.metrics import classification_report, roc_auc_score
print(classification_report(y_test, ytest))
```

```python
print("AUC = ",roc_auc_score(y_test,ytest))
```
