## EEG files into list

```python
def load_eeg_files(path): 
    fichiers=[] 
    for dirname, _, filenames in os.walk(path): 
        for filename in filenames: 
            fichiers.append(os.path.join(dirname, filename)) 
    edflist=[]
    for f in fichiers:
        if f.endswith(".edf"):
            edflist.append(f)
    
    return edflist
```
